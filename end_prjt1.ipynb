#### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
emp_count=df["Team"].value_counts()
tot_players=len(df)
emp_percent=(emp_count/tot_players)*100
result_df1=pd.DataFrame({"No:of players":emp_count,"percentage splitting":emp_percent})
result_df1
```

#### 2.Segregate the employees w.r.t different positions.

```python
grouped_by_position = df.groupby('Position')
for position, group in grouped_by_position:
    print(f"Employees in {position} position:")
    for name in group['Name']:
        print(name)
    print("\n")
```

#### 3.Find from which age group most of the employees belong to.

```python

age_group=pd.cut(df["Age"],bins=[20,25,30,35,40],labels=['20-24','25-29','30-34','35-40'])
age_group_counts=age_group.value_counts()
age_group_counts
```

```python
most_common_age_group=age_group_counts.idxmax()
no_of_emp=age_group_counts.max()
print(f"The age group most of the employees belongs to is {most_common_age_group} with {no_of_emp} employees")
```

#### 4. Find out under which team and position , spending in terms of salary is high

##### Most spending team

```python

group_by_team=df.groupby('Team')['Salary'].max()
group_by_team
```

```python
team=group_by_team.idxmax()
print(f"The team spending highest on salary is {team}")
```

##### Most spend and position

```python
group_by_pos=df.groupby('Position')['Salary'].max()
group_by_pos
```

```python
position=group_by_pos.idxmax()
print(f"Teams spend most on {position} position")
```

#### 5.Find if there is any correlation between age and salary , represent it visually.

```python
#Correlation between age and salary
corr=df['Age'].corr(df['Salary'])
print("corelation between age and salary is",corr)
```

```python
sns.scatterplot(x='Age',y='Salary',data=df)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary of Employees')
plt.grid(True)
plt.show()
```
